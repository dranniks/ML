# Analysis data

```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')

print('Columns: ', train_data.columns.tolist(), '\n')

print(train_data.info(), '\n')

print(train_data.isnull().sum())

print(train_data.head())
```

There are many missing values in the data. Handle the missing values.

```python
filled_train_data = train_data
filled_train_data['Age'] = filled_train_data.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))

filled_train_data['Has_Cabin'] = filled_train_data['Cabin'].notnull().astype(int)
filled_train_data = filled_train_data.drop('Cabin', axis=1)

mode_embarked = filled_train_data['Embarked'].mode()[0]
filled_train_data['Embarked'] = filled_train_data['Embarked'].fillna(mode_embarked)

print(filled_train_data.isnull().sum())
```
